```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.elon.edu/u/news"
```

```python
res = requests.get(url)
```

```python
soup = BeautifulSoup(res.content)
```

```python
soup
```

```python
news_titles = []
for h3 in soup.find_all("h3"):
    news_titles.append(h3.text)
```

```python
news_titles[0:15]
```

```python
soup
```

```python
links = []
for a in soup.find_all("a"):
    links.append(a.get("href"))
# will always be href to be the links 
# 'class' is used as an identifier for a group
```

```python
links_only_for_news = []
for a in soup.find_all("a", "story-item--link"):
    links_only_for_news.append(a.get("href"))
    # would be 'class = story-item--link' but class is already a defined name in python
```

```python
links_only_for_news
```

```python
links
```

```python
#what <a> classes exist in the response
aclasses = []
for a in soup.find_all("a"):
    aclasses.append(a.get("class"))
```

```python
temp =[]
for ac in aclasses:
    #ac is a list
    if ac:
        for x in ac:
            temp.append(x)
    else:
        continue 
```

```python
set(temp)
```

```python
output =[]
for a in soup.find_all("a", "story-item--link"):
    temp = [a.find("h3").text, a.get("href")] # create a list temp
    output.append(temp) # appends output to become a list of lists 
```
